The Python Imaging Library (PIL) provides general image handling and lots of useful basic image operations like resizing, cropping, rotating, color conversion and much more. With PIL, you can read images from most formats and write to the most common ones.

**1. The PIL function open() creates a PIL image object. To read an image, use the following Python Code:**

```python
import matplotlib.pyplot as plt
from PIL import Image 
from pylab import *

Im = Image.open('data/build.jpg')
plt.figure(figsize=(5,7))
plt.imshow(array(Im),cmap='gray')
plt.axis('off')
plt.show()
```

**2. Cropping a region** from an image is done using the **crop() method**. To **convert the cropped image to grayscale**, just add **convert('L')** like this:

```python
box = (1,100,400,400)
region = Im.crop(box).convert('L')
region_array = array(region)
plt.figure(figsize=(5,7))
plt.imshow(region_array,cmap='gray')
plt.axis('off')
plt.show()
```

The **extracted region** can, for example, be **put back using the paste()** method. To **rotate an image**, use counterclockwise angles and **rotate()** like this:

```python
Im.paste(region,box)
Im_rotated = Im.rotate(45)
plt.figure(figsize=(5,7))
plt.imshow(Im_rotated,cmap='gray')
plt.axis('off')
plt.show()
```
