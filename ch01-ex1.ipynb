**1.** Take an image and apply Gaussian blur. Plot the image contours for increasing values of Ïƒ.

With PIL, you can read images from most formats and write to the most common ones. The most important module is the Image module. To read an image, use:

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pylab import *
from numpy import *
from scipy.ndimage import filters

# to read an image and convert it to grayscale, just add convert('L') like this:
Im = array(Image.open('data/build.jpg').convert('L')) 

# show the image
plt.figure(figsize=(9,6))
plt.axis('off')
plt.imshow(Im,cmap='gray')
plt.show()
```

SciPy comes with a module for filtering called **scipy.ndimage.filters** that can be used to compute these convolutions using a fast 1D separation. This exercise needs grayscale images, because the contours need to be taken on a single value for every coordinate [x, y]. 

Now we show the image blurred with increasing *sigma*:

```python
i  = 1;
plt.figure(figsize=(11,9))
for sigma in xrange(0,10,2):
    Im_g = filters.gaussian_filter(Im,sigma)
    plt.subplot(2,5,i)
    plt.imshow(Im_g,cmap='gray')
    plt.title('sigma=%i' %sigma)
    plt.axis('off')
    plt.subplot(2,5,i+5)
    plt.contour(Im_g, origin='image',cmap='gray')
    plt.axis('off')
    i = i + 1;
plt.show()
```

**Note that larger values of *sigma* give less detail !**
